## Analysis (a)
Answering the questions with only scatter plots:

Do students who enroll in more popular majors make more money?
<ul><li>No. Graph 1 indicates that most population sizes per major category cluster around a median salary of 40,000. However, it also indicates that some major categories with lower populations yield much higher median salaries than any major category with higher populations! So, there is a slightly greater possibility of earning a higher salary in a less popular major category than one more popular.</li></ul>
Do students that major in subjects that are majority female make more money?
<ul><li>No. Quite the opposite. Graph 7 actually indicates that, generally, the higher the percentage of women in a major category, the lower the median salary is for the graduates of that major. </li></ul>
Does a link exist between the number full-time employed and median salary?
<ul><li>Yes. There is a slightly negative correlation. Generally speaking, Graph 3 indicates that the more people that are fully employed after their particular major, the lower the median salary. The major categories with the highest value for median salary also have some of the lowest number of full-time employed graduates.</li></ul>

## Conclusion
Answering 3 questions about the dataset with different visualization techniques illustrates the pros and cons of each visualization technique. The scatter plots are easy to read and concisely display a large amount of information, but only between 2 variables. The scatter matrices can illustrate information among 3 variables in a readable way, so complex patterns can be more easily identified. The bar plots are very easy to read but only by assuming the use of a limited number of bars; yet by limiting the number of bars, the conclusions drawn can be misleading and contradict more precise visualization techniques. 

Upon answering the 3 questions, (Q1) whether students enrolled in popular majors make more money and (Q3) whether a link exists between the number of full-time employed graduates and median salary had consistent answers between the first 2 visualization techniques, but differed in the 3rd. Since the data was limited in the last visualization, the more accurate answer in the conlusion from the first 2 visualization analyses: (A1) less popular majors may yield higher median salaries, and (A3) a major with a lower median salary tends to generate a higher number of full-time employed graduates.

Only question 2, (Q2) whether majors with a majority of female students make more money, had a conistent answer among all 3 data visualization techniques. The answer (A2) was unequivocally no. There was a clear negative correlation between share of women per major category and median salary that was no affected by sample size (population) of the major.
